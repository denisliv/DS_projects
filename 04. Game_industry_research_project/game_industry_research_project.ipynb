# Сборный проект: «Исследование игровой индустрии»

В нашем распоряжении исторические данные, до 2016 года, о продажах игр, оценках пользователей и экспертов, жанрах и платформах, полученные из открытых источников.

**Цели:**
1. Подготовить данные к проведению исследования. 
2. Выявить определяющие успешность игры закономерности. 

**Описание данных:**
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию

## Предобработка данных

### Определение и заполнение пропущенных значений

```python
df.loc[df['name'].isna()]
```

Посмотрев на пропущенные значения в столбце `name`, было обнаружено, что пропуски в разрезе признаков `name` и `genre` совпадают. Что немаловажно, для данных игр также отсутствуют значения оценок и рейтинга. С учетом вышесказанного, эти строки следует удалить.

```python
df = df.dropna(axis=0, subset=['name'])
```

```python
df.loc[df['year_of_release'].isna()]
```

```python
df.query('name == "Madden NFL 2004"')
```

Так как в большинстве своем игры выходят одновременно на нескольких платформах, то часть пропущенных значений столбца `year_of_release` можно заполнить на основании года выпуска игры на других платформах.

```python
df['year_of_release'] = df.groupby(['name'])['year_of_release'].transform(lambda x: x.fillna(x.median()))
df = df.dropna(axis=0, subset=['year_of_release'])
```

```python
df.loc[df['rating'].isna()].sample(10)
```

```python
df['rating'].value_counts()
```

Пропущенные значения в столбце `rating` — рейтинг от организации ESRB, обозначающий подходящую возрастную категорию, нельзя заполнить на основании имеющихся данных. Таким образом все пропуски обозначим как *unknown*. Также, судя по общедоступной информации, значение рейтинга *'K-A'* в 1998 году было переименовано организацией в *'Е'*.

```python
df['rating'] = df['rating'].replace('K-A', 'E')
df['rating'] = df['rating'].fillna('unknown')
```

С учетом большого количества пропущенных значений в столбцах `critic_score`	и `user_score`,  а также с учетом того, что замена данных пропусков может привести к смещению финальных оценок, пропущенные значение оставим как *nan*

```python
isna_after = df.isna().sum().sum()
print(f'Всего пропущенных значений {isna_after}.')
print()
na_structure(df)
```

### Изменение типа данных

```python
df['user_score'] = df['user_score'].replace('tbd', np.nan)
```

```python
df['year_of_release'] = df['year_of_release'].astype('uint')

df['user_score'] = df['user_score'].astype('float')

converted_float = df.select_dtypes(include=['float64']).apply(pd.to_numeric, downcast='float')
for column in converted_float.columns:
    df[column] = converted_float[column]

for column in ('genre', 'rating'):
    df[column] = df[column].astype('category')
```

```python
df.info(memory_usage='deep')
```

### Добавление нового признака

```python
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
```

#### Определение актуальных игровых платформ

```python
platform_pv = df.pivot_table(index='platform', values='total_sales', aggfunc=['count', 'sum'])
platform_pv.columns = ['games_count', 'total_sales']
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(200, 100))
ax = platform_pv.sort_values('games_count').plot(y='games_count', kind='barh', figsize=(18, 8), 
                      ec='black', legend=False, fontsize=12, ax=axes[0])
ax.set_title('Распределение количества выпущенных игр по платформам', fontsize=16)
ax.set_xlabel('Количество выпущенных игр, шт.', fontsize=13)
ax.set_ylabel('Платформа', fontsize=13)

ax = platform_pv.sort_values('total_sales').plot(y='total_sales', kind='barh', figsize=(18, 8), 
                      ec='black', legend=False, fontsize=12, color='orangered', ax=axes[1])
ax.set_title('Распределение количества проданных копий по платформам', fontsize=16)
ax.set_xlabel('Количество проданных копий, млн.шт.', fontsize=13)
ax.set_ylabel('Платформа', fontsize=13);
```

```python
platform_pv.sort_values('total_sales', ascending=False).head(10)
```

Отталкиваясь от данных можно заметить, что как по абсолютному количеству выпущенных игр, так и по количеству проданных копий лидером является платформа *PS2*. За доступный период, всего для нее было проданно более *1200 млн.копий* и выпущено более *2000* наименований игр.  Также по обоим этим параметрам в 5 лидеров входят платформы *X360*, *PS3*, *Wii*, *DS*. Также можно выделить платформы *PS* и *PS4*. Первая в свое время поспособствовала развитию игровой индустрии, появившись в *1994 году* и стала прародителем линейки, дав лидера *PS2*. Вторая же является логическим продолжением линейки *PS* и до *2016 года*, возможно, еще не успела приобрести такой популярности.

```python
column_platform_pv = platform_pv.sort_values('total_sales', ascending=False).head(7).index
yor_platform_pv = df.query('year_of_release >= 1994 and platform in @column_platform_pv').pivot_table(index='year_of_release', 
                                                                  columns='platform', values='total_sales', 
                                                                  aggfunc='sum')
```

```python
ax = yor_platform_pv.plot(style='-o', xticks=range(1994, 2017, 1), figsize=(18, 8), fontsize = 13)
ax.legend(fontsize = 15,
          ncol = 2,
          edgecolor = 'b',
          title = 'Платформы',
          title_fontsize = '15'
         )
ax.set_title('График срока актуальности игровой платформы', fontsize=14)
ax.set_xlabel('Год', fontsize=12)
ax.set_ylabel('Количество проданных копий игр, млн.шт', fontsize=13);
```

```python
result = 0
counter = 0
for i in yor_platform_pv.columns:
    result += yor_platform_pv[yor_platform_pv[i] != 0][i].count()
    counter += 1
print(f'Среднее время актуальности игровой платформы составляет {int(result/counter)} лет')
```

Для достижени поставленной цели — определение закономерностей успешности игры для прогнозирования продаж в настоящем и будущем времени — считаю необходимым отобрать данные только за период *2005-2016 годы*, то есть период наибольшего развития и популярности игровой индустрии, а также с учетом срока жизни игровых платформ.

```python
top_platform = df.query('year_of_release >= 2005').pivot_table(index='platform', 
                                                columns='year_of_release', values='total_sales', 
                                                aggfunc='sum')

fig, axis = plt.subplots(nrows=4, ncols=3, figsize=(20, 20))

plt.suptitle('Доля рынка занимаемая игровыми платформами за период 2005-2016 гг.', fontsize=14)

r, c = 0, 0

for i in top_platform.columns:

    ax = top_platform[top_platform[i] > 5].plot(y=i, kind='pie',
                                                 #shadow=True,
                                                 autopct='%1.1f%%',
                                                 fontsize=14,
                                                 ax=axis[r, c],
                                                 legend=False)
    
    ax.set_title(str(i) + ' год', fontsize=14)
    ax.set_ylabel('')

    
    if c != 2:
        c += 1
    else:
        c = 0
        r += 1
        
plt.tight_layout();
```

Анализируя полученные графики, можно заметить, что начиная с *2013 года* рост относительных продаж отмечался только среди платформ *PS4*, *XOne*. Остальные платформы со временем исчезли с рынка и к *2016 году* только *3DS* и *PC* могли немного конкурировать с абсолютными лидерами.

```python
profitable_platforms = ['PS4', 'XOne', 'PC', '3DS']
top_platform = df.query('year_of_release >= 2013 and platform in @profitable_platforms')
```

```python
plt.figure(figsize=(18, 8))
sns.set(font_scale = 1.4)
ax = sns.boxplot(data=top_platform, x='platform', y='total_sales')
ax.set_title('Диаграмма размаха глобальных продаж игр в разрезе актуальных платформ')
ax.set_ylabel('Количество копий, млн. шт.')
ax.set_xlabel('');
```

```python
top_platform.groupby('platform')['total_sales'].describe()
```

Согласно данным, полученным из описательной статистики и графиков боксплот по топовым платформам за *2013-2016 гг.*, можно сделать вывод, что основные продажи по всем платформам приходятся на небольшое количество игр-бестселлеров, тогда как основная масса наименований (*75 %*) сосредоточена в диапазоне от *0.01 до 0.28*, *0.23*, *0.73*, *0.69 млн. копий* для платформ *3DS*, *PC*, *PS4* и *XOne*, соответственно.

## Проверка гипотез

**Необходимо проверить следующие гипотезы:**
1. Средние пользовательские рейтинги платформ *Xbox One* и *PC* не различаются.
2. Средние пользовательские рейтинги жанров *Action* и *Sports* различаются.

### Гипотеза №1

**H0** — cредние пользовательские рейтинги платформ *Xbox One* и *PC* не различаются.

**H1** — cредние пользовательские рейтинги платформ *Xbox One* и *PC* различаются.

Так как игровая платформа *Xbox One* появилась только в *2013 году*, то проведем два сравнения: 
-  используем оценки пользователей для *PC* начиная с *2013 года*;
-  используем оценки пользователей для *PC* за весь период.

```python
df['user_score'] = df['user_score'].fillna(-1)

x_one = df.query('platform == "XOne" and user_score >= 0')['user_score']
pc_2013 = df.query('platform == "PC" and year_of_release >= 2013 and user_score >= 0')['user_score']                
pc = df.query('platform == "PC" and user_score >= 0')['user_score']  

print(f'Средняя оценка для платформы "Xbox One": {round(x_one.mean(), 2)}')
print(f'Средняя оценка для "PC" начиная с 2013 года: {round(pc_2013.mean(), 2)}')
print(f'Средняя оценка для "PC" за весь период: {round(pc.mean(), 2)}')
print('----------------------------------------------------')

alpha = 0.01

results_2013 = st.ttest_ind(x_one, pc_2013, equal_var=False)
results = st.ttest_ind(x_one, pc, equal_var=False)

print('С учетом данных для "PC" начиная с 2013 года', end='\n\n')
print(f'p-value: {results_2013.pvalue}')

if results_2013.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не удалось отвергнуть нулевую гипотезу")
    
print('----------------------------------------------------')
print('С учетом данных для "PC" за весь период', end='\n\n')    
print(f'p-value: {results.pvalue}')

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не удалось отвергнуть нулевую гипотезу")
```

**Вывод:** если сравнивать результаты только за период *2013-2016 гг.* (период существования *Xbox One*), полученные средние значения оценок пользователей для игр платформ *Xbox One* и *PC* статистически значимо не отличаются с p-value > 0.05. В свою очередь, если сравнить оценки пользователей за весь период, то средняя оценка для игр *PC* статистически значимо выше, чем аналогичная оценка для платформы *Xbox One* с p-value < 0.01. Это говорит о том, что игры, появившиеся на заре становления *PC* имеют более высокий пользовательский рейтинг, что и формирует полученную разницу.

### Гипотеза №2

**H0** — cредние пользовательские рейтинги жанров *Action* и *Sports* не различаются.

**H1** — cредние пользовательские рейтинги жанров *Action* и *Sports* различаются.

```python
action = df.query('genre == "Action" and user_score >= 0 and year_of_release >= 2013')['user_score']
sports = df.query('genre == "Sports" and user_score >= 0 and year_of_release >= 2013')['user_score']          
                 
print(f'Средняя оценка для жанра "Action": {round(action.mean(), 2)}')
print(f'Средняя оценка для жанра "Sports": {round(sports.mean(), 2)}', end='\n\n')

alpha = 0.01

results = st.ttest_ind(action, sports, equal_var=False)

print(f'p-value: {results.pvalue}')

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не удалось отвергнуть нулевую гипотезу")
```

**Вывод:** полученные средние значения оценок пользователей для игр жанров *Action* и *Sports* статистически значимо отличаются с p-value < 0.01. Геймеры, в своем большинстве, склонны выше оценивать игры жанра *Action*, чем *Sports*. Это может быть связанно с тем, что по своей сути игры последнего являются симуляторами реальных спортивных игр и ожидания игроков реалистичности визуальной части, физики игры не совпадают с возможностями разработчиков.
